```python
import requests
import bs4
```

Чтобы продемонстрировать, что вы разобрались, как использовать requests и beautiful soup, распарсите: а) из статьи https://en.wikipedia.org/wiki/Bias-variance_tradeoff все заголовки верхнего уровня; б) со страницы https://en.wikipedia.org/wiki/Category:Machine_learning_algorithms названия всех статей в категории Machine Learning Algorithms

а) все заголовки верхнего уровня, имеющие смысл для статьи

```python
req1 = requests.get('https://en.wikipedia.org/wiki/Bias–variance_tradeoff')
```

```python
parser1 = bs4.BeautifulSoup(req1.text, 'lxml')
```

```python
print(parser1.prettify())
```

# Заголовки уровня h1

```python
y = parser1.select("h1")
for result in y:
    print(result.text)
    print("\n------\n")
```

## Заголовки уровня h2

```python
y1 = parser1.select(".mw-headline")
for result in y1:
    print(result.text)
    print("\n------\n")
```

б) Вывести названия статей по ML

```python
req2 = requests.get('https://en.wikipedia.org/wiki/Category:Machine_learning_algorithms ')
parser2 = bs4.BeautifulSoup(req2.text, 'lxml')
print(parser2.prettify())
```

Выводим необходимые темы раздела, используя CSS селектор.

```python
pages = parser2.select(".mw-category-group > ul > li")
for page in pages:
    print(page.text)
    print("\n------\n")
```
